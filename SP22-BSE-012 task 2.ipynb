```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,3,4,5])
ypoints = np.array([2, 4, 6, 8, 10])

plt.plot(xpoints, ypoints)
plt.title("Line Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")


plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
ypoints = np.array([99, 86, 87, 88, 100, 86, 103, 87, 94, 78])

plt.scatter(xpoints, ypoints, s=50, c='pink', alpha=1)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(['A', 'B', 'C', 'D', 'E'])
y = np.array([5, 7, 3, 8, 6])

plt.bar(x, y, color=['lavender'], alpha=1)
plt.show()
```
